```python
import numpy as np
import scipy as sc
from numpy import format_float_positional as ffp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, fbeta_score, roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from time import perf_counter
from tqdm import tqdm
from joblib import Parallel, delayed

sns.set_theme(context='notebook', style='ticks')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'cm'
```

# Funções e Classes

```python
def evaluate(model, X_train, X_test, y_train, y_test,
             threshold=0.5, beta_fscore=1, dig=3, model_title='', threshold_step=0.01,
             display_metrics_table=True, plot_confusion_matrix=True, plot_roc_prc=True):
    y_train_prob_pred = model.predict_proba(X=X_train)[:, 1]
    y_test_prob_pred = model.predict_proba(X=X_test)[:, 1]
    y_train_pred = np.int32(y_train_prob_pred >= threshold)
    y_test_pred = np.int32(y_test_prob_pred >= threshold)

    threshold_vec = np.arange(0., 1. + 0.01, 0.01)[::-1]
    fpr_vec = np.empty_like(threshold_vec)
    tpr_vec = np.empty_like(threshold_vec)
    pre_vec = np.empty_like(threshold_vec)
    for i, thr in enumerate(threshold_vec):
        y_pred = np.int32(y_test_prob_pred >= thr)
        conf_matrix = confusion_matrix(y_pred=y_pred, y_true=y_test)
        fpr, tpr = conf_matrix[:, 1] / conf_matrix.sum(axis=1)
        fpr_vec[i] = fpr
        tpr_vec[i] = tpr
        pre_vec[i] = precision_score(y_pred=y_pred, y_true=y_test, zero_division=1.)
    
    acc_train = accuracy_score(y_pred=y_train_pred, y_true=y_train)
    pre_train = precision_score(y_pred=y_train_pred, y_true=y_train)
    rec_train = recall_score(y_pred=y_train_pred, y_true=y_train)
    f1s_train = f1_score(y_pred=y_train_pred, y_true=y_train)
    fbs_train = fbeta_score(y_pred=y_train_pred, y_true=y_train, beta=beta_fscore)
    
    acc_test = accuracy_score(y_pred=y_test_pred, y_true=y_test)
    pre_test = precision_score(y_pred=y_test_pred, y_true=y_test)
    rec_test = recall_score(y_pred=y_test_pred, y_true=y_test)
    f1s_test = f1_score(y_pred=y_test_pred, y_true=y_test)
    fbs_test = fbeta_score(y_pred=y_test_pred, y_true=y_test, beta=beta_fscore)
    auroc_test = sc.integrate.trapezoid(x=fpr_vec, y=tpr_vec)
    auprc_test = sc.integrate.trapezoid(x=tpr_vec, y=pre_vec)

    metrics_dict = {
        'Treinamento':[ffp(acc_train, dig, min_digits=dig), ffp(pre_train, dig, min_digits=dig), ffp(rec_train, dig, min_digits=dig),
                       ffp(f1s_train, dig, min_digits=dig), ffp(fbs_train, dig, min_digits=dig),
                       '-----', '-----'],
        'Teste':[ffp(acc_test, dig, min_digits=dig), ffp(pre_test, dig, min_digits=dig), ffp(rec_test, dig, min_digits=dig),
                 ffp(f1s_test, dig, min_digits=dig), ffp(fbs_test, dig, min_digits=dig),
                 ffp(auroc_test, dig, min_digits=dig), ffp(auprc_test, dig, min_digits=dig)]
    }

    if display_metrics_table:
        display(
            pd.DataFrame(metrics_dict,
                index=['Acurácia','Precisão','Recall','F1-Score','F$_{\\beta}$-Score','AUROC', 'AUPRC']
                        ).style.set_caption('Métricas de Avaliação'+model_title)
                )

    if plot_confusion_matrix:
        conf_matrix = confusion_matrix(y_pred=y_test_pred, y_true=y_test)
        conf_matrix_normalized_pcol = conf_matrix / conf_matrix.sum(axis=0)
        conf_matrix_normalized_prow = conf_matrix / conf_matrix.sum(axis=1).reshape(-1, 1)
        conf_matrix_normalized = conf_matrix / conf_matrix.sum()
        
        fig, ax = plt.subplots(1, 4, figsize=(2.5 * 4, 2.5), layout='constrained')
        sns.heatmap(conf_matrix, ax=ax[0], annot=True, annot_kws={'size':15}, fmt='.0f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01)
        sns.heatmap(conf_matrix_normalized_prow, ax=ax[1], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        sns.heatmap(conf_matrix_normalized_pcol, ax=ax[2], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        sns.heatmap(conf_matrix_normalized, ax=ax[3], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        for i in range(4):
            ax[i].set_xticklabels(['No', 'Yes'])
            ax[i].set_yticklabels(['No', 'Yes'], rotation=0)
            ax[i].set_xlabel('Pred.')
            ax[i].set_ylabel('Obs.')
        ax[0].set_title('Absoluta')
        ax[1].set_title('Relativa por linhas')
        ax[2].set_title('Relativa por colunas')
        ax[3].set_title('Relativa total')
        fig.suptitle('Matrizes de Confusão (Teste)'+model_title)
        plt.show()

    if plot_roc_prc:
        threshold_vec = np.arange(0., 1. + threshold_step, threshold_step)[::-1]
        fpr_vec = np.empty_like(threshold_vec)
        tpr_vec = np.empty_like(threshold_vec)
        pre_vec = np.empty_like(threshold_vec)
        for i, thr in enumerate(threshold_vec):
            y_pred = np.int32(y_test_prob_pred >= thr)
            conf_matrix = confusion_matrix(y_pred=y_pred, y_true=y_test)
            fpr, tpr = conf_matrix[:, 1] / conf_matrix.sum(axis=1)
            fpr_vec[i] = fpr
            tpr_vec[i] = tpr
            pre_vec[i] = precision_score(y_pred=y_pred, y_true=y_test, zero_division=1.)
            
        actual_threshold_id = np.argmin(np.abs(tpr_vec - rec_test))

        fig, ax = plt.subplots(1, 2, figsize=(5 * 2, 5), layout='constrained')        
        sns.lineplot(ax=ax[0], x=fpr_vec, y=tpr_vec, lw=1.5, color='black')
        sns.lineplot(ax=ax[0], x=[0., 1.], y=[0., 1.], lw=.7, ls='--', color='black')
        sns.lineplot(ax=ax[1], x=tpr_vec, y=pre_vec, lw=1.5, color='black')
        sns.lineplot(ax=ax[1], x=[0., 1.], y=[y_test.mean()] * 2, lw=.7, ls='--', color='black')
        sns.scatterplot(ax=ax[0], x=[fpr_vec[actual_threshold_id].item()], y=[tpr_vec[actual_threshold_id].item()],
                        marker='o', color='white', edgecolor='black', lw=1.,
                        label=f'Actual ({threshold_vec[actual_threshold_id].item():.2f})', zorder=3)
        sns.scatterplot(ax=ax[1], x=[tpr_vec[actual_threshold_id].item()], y=[pre_vec[actual_threshold_id].item()],
                        marker='o', color='white', edgecolor='black', lw=1.,
                        label=f'Actual ({threshold_vec[actual_threshold_id.item()]:.2f})', zorder=3)
        ax[0].fill_between(x=fpr_vec, y1=np.zeros_like(tpr_vec), y2=tpr_vec, color='black', alpha=.05)
        ax[1].fill_between(x=tpr_vec, y1=np.zeros_like(tpr_vec), y2=pre_vec, color='black', alpha=.05)
        ax[0].set_xticks(np.arange(0, 1.01, .1))
        ax[0].set_yticks(np.arange(0, 1.01, .1))
        ax[1].set_xticks(np.arange(0, 1.01, .1))
        ax[1].set_yticks(np.arange(0, 1.01, .1))
        ax[0].set_xlim(0. - 0.01, 1. + 0.01)
        ax[0].set_ylim(0. - 0.01, 1. + 0.01)
        ax[1].set_xlim(0. - 0.01, 1. + 0.01)
        ax[1].set_ylim(y_test.mean() - .01, 1. + 0.01)
        ax[0].set_xlabel('FPR')
        ax[0].set_ylabel('TPR')
        ax[1].set_xlabel('Recall')
        ax[1].set_ylabel('Precisão')
        ax[0].set_title('ROC'+model_title+'\n$(\\text{AUROC}='+ffp(auroc_test, dig, min_digits=dig)+')$')
        ax[1].set_title('PRC'+model_title+'\n$(\\text{AUPRC}='+ffp(auprc_test, dig, min_digits=dig)+')$')
        ax[0].grid(lw=.5)
        ax[1].grid(lw=.5)
        ax[0].legend(loc='best')
        ax[1].legend(loc='best')
        plt.show()

    return metrics_dict

def evaluate_nn_torch(model, train_loader, val_loader, test_loader,
                      threshold=0.5, beta_fscore=1, dig=3, model_title='', threshold_step=0.01,
                      display_metrics_table=True, plot_confusion_matrix=True, plot_roc_prc=True):
    y_train_prob_pred = torch.tensor([])
    y_val_prob_pred = torch.tensor([])
    y_test_prob_pred = torch.tensor([])
    y_train = torch.tensor([])
    
    model.eval()
    with torch.no_grad():
        for X_batch, y_batch in tqdm(train_loader, desc='Avaliação no dataset de TREINAMENTO'):
            y_train_prob_pred = torch.concatenate((y_train_prob_pred, model(X_batch)))
            y_train = torch.concatenate((y_train, y_batch))
            
        for X_batch, _ in tqdm(val_loader, desc='Avaliação no dataset de VALIDAÇÃO'):
            y_val_prob_pred = torch.concatenate((y_val_prob_pred, model(X_batch)))

        for X_batch, _ in tqdm(test_loader, desc='Avaliação no dataset de TESTE'):
            y_test_prob_pred = torch.concatenate((y_test_prob_pred, model(X_batch)))
            
    y_train_pred = (y_train_prob_pred >= threshold).to(torch.float32)
    y_val_pred = (y_val_prob_pred >= threshold).to(torch.float32)
    y_test_pred = (y_test_prob_pred >= threshold).to(torch.float32)

    threshold_vec = torch.arange(1., 0. - 0.01, -0.01)
    fpr_vec = torch.empty_like(threshold_vec)
    tpr_vec = torch.empty_like(threshold_vec)
    pre_vec = torch.empty_like(threshold_vec)
    for i, thr in enumerate(threshold_vec):
        y_pred = (y_test_prob_pred >= thr).to(torch.float32)
        conf_matrix = confusion_matrix(y_pred=y_pred, y_true=test_loader.dataset.y)
        fpr, tpr = conf_matrix[:, 1] / conf_matrix.sum(axis=1)
        fpr_vec[i] = fpr
        tpr_vec[i] = tpr
        pre_vec[i] = precision_score(y_pred=y_pred, y_true=y_test, zero_division=1.)
    
    acc_train = accuracy_score(y_pred=y_train_pred, y_true=y_train)
    pre_train = precision_score(y_pred=y_train_pred, y_true=y_train)
    rec_train = recall_score(y_pred=y_train_pred, y_true=y_train)
    f1s_train = f1_score(y_pred=y_train_pred, y_true=y_train)
    fbs_train = fbeta_score(y_pred=y_train_pred, y_true=y_train, beta=beta_fscore)

    acc_val = accuracy_score(y_pred=y_val_pred, y_true=val_loader.dataset.y)
    pre_val = precision_score(y_pred=y_val_pred, y_true=val_loader.dataset.y)
    rec_val = recall_score(y_pred=y_val_pred, y_true=val_loader.dataset.y)
    f1s_val = f1_score(y_pred=y_val_pred, y_true=val_loader.dataset.y)
    fbs_val = fbeta_score(y_pred=y_val_pred, y_true=val_loader.dataset.y, beta=beta_fscore)
    
    acc_test = accuracy_score(y_pred=y_test_pred, y_true=test_loader.dataset.y)
    pre_test = precision_score(y_pred=y_test_pred, y_true=test_loader.dataset.y)
    rec_test = recall_score(y_pred=y_test_pred, y_true=test_loader.dataset.y)
    f1s_test = f1_score(y_pred=y_test_pred, y_true=test_loader.dataset.y)
    fbs_test = fbeta_score(y_pred=y_test_pred, y_true=test_loader.dataset.y, beta=beta_fscore)
    auroc_test = sc.integrate.trapezoid(x=fpr_vec, y=tpr_vec)
    auprc_test = sc.integrate.trapezoid(x=tpr_vec, y=pre_vec)

    metrics_dict = {
        'Treinamento':[ffp(acc_train, dig, min_digits=dig), ffp(pre_train, dig, min_digits=dig), ffp(rec_train, dig, min_digits=dig),
                       ffp(f1s_train, dig, min_digits=dig), ffp(fbs_train, dig, min_digits=dig),
                       '-----', '-----'],
        'Validação':[ffp(acc_val, dig, min_digits=dig), ffp(pre_val, dig, min_digits=dig), ffp(rec_val, dig, min_digits=dig),
                     ffp(f1s_val, dig, min_digits=dig), ffp(fbs_val, dig, min_digits=dig),
                     '-----', '-----'],
        'Teste':[ffp(acc_test, dig, min_digits=dig), ffp(pre_test, dig, min_digits=dig), ffp(rec_test, dig, min_digits=dig),
                 ffp(f1s_test, dig, min_digits=dig), ffp(fbs_test, dig, min_digits=dig),
                 ffp(auroc_test, dig, min_digits=dig), ffp(auprc_test, dig, min_digits=dig)]
    }

    if display_metrics_table:
        display(
            pd.DataFrame(metrics_dict,
                index=['Acurácia','Precisão','Recall','F1-Score','F$_{\\beta}$-Score','AUROC', 'AUPRC']
                        ).style.set_caption('Métricas de Avaliação'+model_title)
                )

    if plot_confusion_matrix:
        conf_matrix = confusion_matrix(y_pred=y_test_pred, y_true=y_test)
        conf_matrix_normalized_pcol = conf_matrix / conf_matrix.sum(axis=0)
        conf_matrix_normalized_prow = conf_matrix / conf_matrix.sum(axis=1).reshape(-1, 1)
        conf_matrix_normalized = conf_matrix / conf_matrix.sum()
        
        fig, ax = plt.subplots(1, 4, figsize=(2.5 * 4, 2.5), layout='constrained')
        sns.heatmap(conf_matrix, ax=ax[0], annot=True, annot_kws={'size':15}, fmt='.0f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01)
        sns.heatmap(conf_matrix_normalized_prow, ax=ax[1], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        sns.heatmap(conf_matrix_normalized_pcol, ax=ax[2], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        sns.heatmap(conf_matrix_normalized, ax=ax[3], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        for i in range(4):
            ax[i].set_xticklabels(['No', 'Yes'])
            ax[i].set_yticklabels(['No', 'Yes'], rotation=0)
            ax[i].set_xlabel('Pred.')
            ax[i].set_ylabel('Obs.')
        ax[0].set_title('Absoluta')
        ax[1].set_title('Relativa por linhas')
        ax[2].set_title('Relativa por colunas')
        ax[3].set_title('Relativa total')
        fig.suptitle('Matrizes de Confusão (Teste)'+model_title)
        plt.show()

    if plot_roc_prc:
        threshold_vec = torch.arange(1., 0. - threshold_step, -threshold_step)
        fpr_vec = torch.empty_like(threshold_vec)
        tpr_vec = torch.empty_like(threshold_vec)
        pre_vec = torch.empty_like(threshold_vec)
        for i, thr in enumerate(threshold_vec):
            y_pred = (y_test_prob_pred >= thr).to(torch.float32)
            conf_matrix = confusion_matrix(y_pred=y_pred, y_true=y_test)
            fpr, tpr = conf_matrix[:, 1] / conf_matrix.sum(axis=1)
            fpr_vec[i] = fpr
            tpr_vec[i] = tpr
            pre_vec[i] = precision_score(y_pred=y_pred, y_true=y_test, zero_division=1.)
            
        actual_threshold_id = torch.argmin(torch.abs(threshold_vec - threshold))

        fig, ax = plt.subplots(1, 2, figsize=(5 * 2, 5), layout='constrained')        
        sns.lineplot(ax=ax[0], x=fpr_vec, y=tpr_vec, lw=1.5, color='black')
        sns.lineplot(ax=ax[0], x=[0., 1.], y=[0., 1.], lw=.7, ls='--', color='black')
        sns.lineplot(ax=ax[1], x=tpr_vec, y=pre_vec, lw=1.5, color='black')
        sns.lineplot(ax=ax[1], x=[0., 1.], y=[y_test.mean()] * 2, lw=.7, ls='--', color='black')
        sns.scatterplot(ax=ax[0], x=[fpr_vec[actual_threshold_id].item()], y=[tpr_vec[actual_threshold_id].item()],
                        marker='o', color='white', edgecolor='black', lw=1.,
                        label=f'Actual ({threshold_vec[actual_threshold_id].item():.2f})', zorder=3)
        sns.scatterplot(ax=ax[1], x=[tpr_vec[actual_threshold_id].item()], y=[pre_vec[actual_threshold_id].item()],
                        marker='o', color='white', edgecolor='black', lw=1.,
                        label=f'Actual ({threshold_vec[actual_threshold_id.item()]:.2f})', zorder=3)
        ax[0].fill_between(x=fpr_vec, y1=torch.zeros_like(tpr_vec), y2=tpr_vec, color='black', alpha=.05)
        ax[1].fill_between(x=tpr_vec, y1=torch.zeros_like(tpr_vec), y2=pre_vec, color='black', alpha=.05)
        ax[0].set_xticks(np.arange(0, 1.01, .1))
        ax[0].set_yticks(np.arange(0, 1.01, .1))
        ax[1].set_xticks(np.arange(0, 1.01, .1))
        ax[1].set_yticks(np.arange(0, 1.01, .1))
        ax[0].set_xlim(0. - 0.01, 1. + 0.01)
        ax[0].set_ylim(0. - 0.01, 1. + 0.01)
        ax[1].set_xlim(0. - 0.01, 1. + 0.01)
        ax[1].set_ylim(y_test.mean() - .01, 1. + 0.01)
        ax[0].set_xlabel('FPR')
        ax[0].set_ylabel('TPR')
        ax[1].set_xlabel('Recall')
        ax[1].set_ylabel('Precisão')
        ax[0].set_title('ROC'+model_title+'\n$(\\text{AUROC}='+ffp(auroc_test, dig, min_digits=dig)+')$')
        ax[1].set_title('PRC'+model_title+'\n$(\\text{AUPRC}='+ffp(auprc_test, dig, min_digits=dig)+')$')
        ax[0].grid(lw=.5)
        ax[1].grid(lw=.5)
        ax[0].legend(loc='best')
        ax[1].legend(loc='best')
        plt.show()

    return metrics_dict

class BinaryRandomGuess:
    def __init__(self, threshold=0.5, seed=None):
        self.threshold = threshold
        self._uniform = sc.stats.uniform(0, 1)
        self._seed = seed

    def fit(self, X, y):
        pass

    def predict(self, X):
        n = X.shape[0]
        y_pred = self._uniform.rvs(size=n, random_state=self._seed)
        y_pred = np.int32(y_pred >= self.threshold)
        return y_pred

    def predict_proba(self, X):
        n = X.shape[0]
        y_prob_pred = self._uniform.rvs(size=n, random_state=self._seed).reshape(-1, 1)
        y_prob_pred = np.concatenate((1. - y_prob_pred, y_prob_pred), axis=1)
        return y_prob_pred

class MLP(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_layers, activations, dropout,
                 weight_init_fn=lambda W:W, bias_init_fn=lambda b:b, bias_init_factor=1.,
                 seed=None):
        super().__init__()
        torch.manual_seed(seed = seed)

        self.n_hidden_layers = len(hidden_layers)
        self.network = nn.Sequential()
        
        self.network.add_module('Inp - H1', nn.Linear(input_dim, hidden_layers[0]))
        with torch.no_grad():
            weight_init_fn(self.network[-1].weight)
            if bias_init_fn.__repr__()[:19] == nn.init.constant_.__repr__()[:19]:
                bias_init_fn(self.network[-1].bias, val=bias_init_factor)
            else:
                bias_init_fn(self.network[-1].bias)
            
        self.network.add_module('Activation H1', activations[0])

        if len(dropout) > 0:
            self.network.add_module('Dropout H1', nn.Dropout(dropout[0]))
        
        for i in range(1, self.n_hidden_layers):
            self.network.add_module('H'+str(i)+' - H'+str(i+1), nn.Linear(hidden_layers[i-1], hidden_layers[i]))
            with torch.no_grad():
                weight_init_fn(self.network[-1].weight)
                if bias_init_fn.__repr__()[:19] == nn.init.constant_.__repr__()[:19]:
                    bias_init_fn(self.network[-1].bias, val=bias_init_factor)
                else:
                    bias_init_fn(self.network[-1].bias)
                
            self.network.add_module('Activation H'+str(i+1), activations[i])

            if len(dropout) > 0:
                self.network.add_module('Dropout H'+str(i+1), nn.Dropout(dropout[i]))
            
        self.network.add_module('H'+str(i+1)+' - Out', nn.Linear(hidden_layers[-1], output_dim))
        with torch.no_grad():
            weight_init_fn(self.network[-1].weight)
            if bias_init_fn.__repr__()[:19] == nn.init.constant_.__repr__()[:19]:
                bias_init_fn(self.network[-1].bias, val=bias_init_factor)
            else:
                bias_init_fn(self.network[-1].bias)

        if len(activations) == len(hidden_layers) + 1:
            self.network.add_module('Activation H'+str(i+2), activations[i+1])

    def forward(self, x):
        return self.network(x)

    def show_network(self):
        print(self.network)

class TorchDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
    
# Retorna uma função que realça as células que contém os n-ésimos valores máximos
def pd_highlight_nthmax(nth_max=1):
    def highlight_max(s, props=""):
        a = np.array([''] * np.size(s), dtype='object')
        a[np.argsort(s.values)[-nth_max]] = props
        return a
    return highlight_max
```

# MLP (`torch`)

```python
X_train_transformed = transformer_X['MLP'].fit_transform(df_X_est)
X_val_transformed = transformer_X['MLP'].transform(df_X_val)
X_test_transformed = transformer_X['MLP'].transform(df_X_test)
y_train = transformer_y.fit_transform(df_y_est)
y_val = transformer_y.transform(df_y_val)
y_test = transformer_y.transform(df_y_test)

train_dataset = TorchDataset(X=X_train_transformed, y=y_train)
val_dataset = TorchDataset(X=X_val_transformed, y=y_val)
test_dataset = TorchDataset(X=X_test_transformed, y=y_test)

torch.manual_seed(seed=7)
train_loader = DataLoader(dataset=train_dataset, shuffle=True, batch_size=32)
val_loader = DataLoader(dataset=val_dataset, shuffle=False, batch_size=512)
test_loader = DataLoader(dataset=test_dataset, shuffle=False, batch_size=512)
```

```python
model['MLP'] = MLP(input_dim=X_train_transformed.shape[1],
                   output_dim=1,
                   hidden_layers=[256, 256],
                   activations=[nn.ReLU()] * 2 + [nn.Sigmoid()],
                   dropout=[],
                   weight_init_fn=nn.init.kaiming_normal_,
                   bias_init_fn=nn.init.constant_,
                   bias_init_factor=0.0001,
                   seed=7)

model['MLP'].show_network()
```

```python
n_epochs = 200
loss_fn = nn.BCELoss()
optimizer = torch.optim.AdamW(params=model['MLP'].parameters(),
                              lr=1e-5,
                              weight_decay=1e-4)

loss_train_list, loss_val_list = [], []
best_val_loss = torch.inf
best_epoch = 0

torch.manual_seed(seed=7)
for epoch in tqdm(range(n_epochs), desc='Treinamento MLP'):
    model['MLP'].train()
    total_train_loss = 0.
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        y_pred = model['MLP'](X_batch)
        loss = loss_fn(y_pred, y_batch)
        loss.backward()
        optimizer.step()
        total_train_loss += loss.item() * X_batch.shape[0]
    loss_train_list.append(total_train_loss / train_dataset.X.shape[0])

    model['MLP'].eval()
    total_val_loss = 0.
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            y_pred = model['MLP'](X_batch)
            loss = loss_fn(y_pred, y_batch)
            total_val_loss += loss.item() * X_batch.shape[0]
        loss_val_list.append(total_val_loss / val_dataset.X.shape[0])
        if loss_val_list[-1] < best_val_loss:
            best_val_loss = loss_val_list[-1]
            best_epoch = epoch + 1
            torch.save(model['MLP'].state_dict(), 'best_nn.pth')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5), layout='constrained')

sns.lineplot(ax=ax, x=range(1, len(loss_train_list) + 1), y=loss_train_list,   lw=1.5, ls='-', color='black', alpha=1., label='Treinamento')
sns.lineplot(ax=ax, x=range(1, len(loss_val_list) + 1), y=loss_val_list, lw=.7, ls='--',  color='black', alpha=1., label='Validação')
sns.scatterplot(ax=ax, x=[best_epoch], y=[best_val_loss], marker='X', color='red', s=60, zorder=2, label='Melhor modelo')

ax.legend()
ax.set_title('Curvas de Treinamento e Validação')
ax.set_ylabel('BCE')
ax.set_xlabel('Épocas')
ax.grid(lw=.5)
plt.show()
```

## Ajuste de *Threshold* visando decisões de negócio

Consideremos:
- A classe de interesse maior, que são os **clientes que cometerão *churn***, é representada por **1**;
- A classe dos que **não cometerão** é representada por **0**;
- O custo de errar a classe **1**, ou seja, **classificar um cliente que é propenso a *churn* como não propenso**, é representado por $\text{c}_{10}$; 
- O custo de errar a classe **0**, ou seja, **classificar um cliente que não é propenso a *churn* como propenso**, é representado por $\text{c}_{01}$;
- O custo $\text{c}_{01}$ está relacionado à **investir na retenção de clientes que NÃO IRÃO ABANDONAR o serviço**, alocando recursos desnecessariamente;
- O custo $\text{c}_{10}$ está relacionado a **não investir na retenção de clientes IRÃO ABANDONAR o serviço**, prejudicando a receita recorrente.
- Ambos são prejudiciais, mas no contexto de negócios, **não investir na retenção de clientes que irão praticar *churn* é mais preocupante**, portanto o custo $\text{c}_{10}$ é maior.


Se um modelo retorna as probabilidades de classificação das classes **1** e **0** como $p_1$ e $p_0=1-p_1$, respectivamente, devemos escolher a classe **1** caso:

$\begin{eqnarray}
&&p_1 \cdot c_{10} > p_0 \cdot c_{01} \Rightarrow \\
&&p_1 \cdot c_{10} > (1-p_1) \cdot c_{01} \Rightarrow \\
&&p_1 > \dfrac{c_{01}}{c_{01}+c_{10}};
\end{eqnarray}$

Se considerarmos que o custo de errar a classe **1** é $R$ vezes o custo de errar a classe **0** ($c_{10}=R \cdot c_{01}$), devemos escolher a classe **1** caso:

$\begin{eqnarray}
&&p_1 > \dfrac{1}{R+1}.
\end{eqnarray}$

#### **Simulação com $R=2$ ($c_{10}$ 2 vezes maior do que $c_{01}$)**
> 
Vamos considerar:
>O **custo de não reter clientes que irão abandonar o serviço** é **2x** maior do que o **custo de investir desnecessariamente em clientes que não abandonariam**;

Isso significa que devemos escolher a classe **1** caso $p_1 > 1/3$, ou seja, **o novo *threshold* será 0.33.**

#### Observações da comparação final:
- Como as ponderações para os diferentes tipos de predições já estão sendo levadas em conta a partir dos diferentes custos de erros, a ponderação pela métrica $\text{F}_{\beta}\text{ -Score}$ ($\beta=2$) agora faz menos sentido;
- O modelo **LR** apresentou o melhor equilíbrio entre precisão e *recall*;
- O modelo **SVM** cresceu significamente em *recall*, ao custo de ter sua precisão reduzida também significativamente;
- Os modelos **DT** e **KNN** apresentaram-se competitivos, com **DT** priorizando precisão, e **KNN** priorizando *recall*;
