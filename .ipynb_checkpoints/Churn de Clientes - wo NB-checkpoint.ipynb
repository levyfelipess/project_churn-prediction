# Funções e Classes

```python
def evaluate(model, X_train, X_test, y_train, y_test,
             threshold=0.5, beta_fscore=1, dig=3, model_title='', threshold_step=0.01,
             display_metrics_table=True, plot_confusion_matrix=True, plot_roc_prc=True):
    y_train_prob_pred = model.predict_proba(X=X_train)[:, 1]
    y_test_prob_pred = model.predict_proba(X=X_test)[:, 1]
    y_train_pred = np.int32(y_train_prob_pred >= threshold)
    y_test_pred = np.int32(y_test_prob_pred >= threshold)

    threshold_vec = np.arange(0., 1. + 0.01, 0.01)[::-1]
    fpr_vec = np.empty_like(threshold_vec)
    tpr_vec = np.empty_like(threshold_vec)
    pre_vec = np.empty_like(threshold_vec)
    for i, thr in enumerate(threshold_vec):
        y_pred = np.int32(y_test_prob_pred >= thr)
        conf_matrix = confusion_matrix(y_pred=y_pred, y_true=y_test)
        fpr, tpr = conf_matrix[:, 1] / conf_matrix.sum(axis=1)
        fpr_vec[i] = fpr
        tpr_vec[i] = tpr
        pre_vec[i] = precision_score(y_pred=y_pred, y_true=y_test, zero_division=1.)
    
    acc_train = accuracy_score(y_pred=y_train_pred, y_true=y_train)
    pre_train = precision_score(y_pred=y_train_pred, y_true=y_train)
    rec_train = recall_score(y_pred=y_train_pred, y_true=y_train)
    f1s_train = f1_score(y_pred=y_train_pred, y_true=y_train)
    fbs_train = fbeta_score(y_pred=y_train_pred, y_true=y_train, beta=beta_fscore)
    
    acc_test = accuracy_score(y_pred=y_test_pred, y_true=y_test)
    pre_test = precision_score(y_pred=y_test_pred, y_true=y_test)
    rec_test = recall_score(y_pred=y_test_pred, y_true=y_test)
    f1s_test = f1_score(y_pred=y_test_pred, y_true=y_test)
    fbs_test = fbeta_score(y_pred=y_test_pred, y_true=y_test, beta=beta_fscore)
    auroc_test = trapezoid(x=fpr_vec, y=tpr_vec)
    auprc_test = trapezoid(x=tpr_vec, y=pre_vec)

    metrics_dict = {
        'Treinamento':[ffp(acc_train, dig, min_digits=dig), ffp(pre_train, dig, min_digits=dig), ffp(rec_train, dig, min_digits=dig),
                       ffp(f1s_train, dig, min_digits=dig), ffp(fbs_train, dig, min_digits=dig),
                       '-----', '-----'],
        'Teste':[ffp(acc_test, dig, min_digits=dig), ffp(pre_test, dig, min_digits=dig), ffp(rec_test, dig, min_digits=dig),
                 ffp(f1s_test, dig, min_digits=dig), ffp(fbs_test, dig, min_digits=dig),
                 ffp(auroc_test, dig, min_digits=dig), ffp(auprc_test, dig, min_digits=dig)]
    }

    if display_metrics_table:
        display(
            pd.DataFrame(metrics_dict,
                index=['Acurácia','Precisão','Recall','F1-Score','F$_{\\beta}$-Score','AUROC', 'AUPRC']
                        ).style.set_caption('Métricas de Avaliação'+model_title)
                )

    if plot_confusion_matrix:
        conf_matrix = confusion_matrix(y_pred=y_test_pred, y_true=y_test)
        conf_matrix_normalized_pcol = conf_matrix / conf_matrix.sum(axis=0)
        conf_matrix_normalized_prow = conf_matrix / conf_matrix.sum(axis=1).reshape(-1, 1)
        conf_matrix_normalized = conf_matrix / conf_matrix.sum()
        
        fig, ax = plt.subplots(1, 4, figsize=(2.5 * 4, 2.5), layout='constrained')
        sns.heatmap(conf_matrix, ax=ax[0], annot=True, annot_kws={'size':15}, fmt='.0f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01)
        sns.heatmap(conf_matrix_normalized_prow, ax=ax[1], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        sns.heatmap(conf_matrix_normalized_pcol, ax=ax[2], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        sns.heatmap(conf_matrix_normalized, ax=ax[3], annot=True, annot_kws={'size':15}, fmt='.3f', cmap='Blues', cbar=False, linecolor='black', linewidths=.01, vmin=0., vmax=1.)
        for i in range(4):
            ax[i].set_xticklabels(['No', 'Yes'])
            ax[i].set_yticklabels(['No', 'Yes'], rotation=0)
            ax[i].set_xlabel('Pred.')
            ax[i].set_ylabel('Obs.')
        ax[0].set_title('Absoluta')
        ax[1].set_title('Relativa por linhas')
        ax[2].set_title('Relativa por colunas')
        ax[3].set_title('Relativa total')
        fig.suptitle('Matrizes de Confusão (Teste)'+model_title)
        plt.show()

    if plot_roc_prc:
        threshold_vec = np.arange(0., 1. + threshold_step, threshold_step)[::-1]
        fpr_vec = np.empty_like(threshold_vec)
        tpr_vec = np.empty_like(threshold_vec)
        pre_vec = np.empty_like(threshold_vec)
        for i, thr in enumerate(threshold_vec):
            y_pred = np.int32(y_test_prob_pred >= thr)
            conf_matrix = confusion_matrix(y_pred=y_pred, y_true=y_test)
            fpr, tpr = conf_matrix[:, 1] / conf_matrix.sum(axis=1)
            fpr_vec[i] = fpr
            tpr_vec[i] = tpr
            pre_vec[i] = precision_score(y_pred=y_pred, y_true=y_test, zero_division=1.)
            
        actual_threshold_id = np.argmin(np.abs(tpr_vec - rec_test))

        fig, ax = plt.subplots(1, 2, figsize=(5 * 2, 5), layout='constrained')        
        sns.lineplot(ax=ax[0], x=fpr_vec, y=tpr_vec, lw=1.5, color='black')
        sns.lineplot(ax=ax[0], x=[0., 1.], y=[0., 1.], lw=.7, ls='--', color='black')
        sns.lineplot(ax=ax[1], x=tpr_vec, y=pre_vec, lw=1.5, color='black')
        sns.lineplot(ax=ax[1], x=[0., 1.], y=[y_test.mean()] * 2, lw=.7, ls='--', color='black')
        sns.scatterplot(ax=ax[0], x=[fpr_vec[actual_threshold_id]], y=[tpr_vec[actual_threshold_id]],
                        marker='o', color='white', edgecolor='black', lw=1.,
                        label=f'Actual ({threshold_vec[actual_threshold_id]:.2f})', zorder=3)
        sns.scatterplot(ax=ax[1], x=[tpr_vec[actual_threshold_id]], y=[pre_vec[actual_threshold_id]],
                        marker='o', color='white', edgecolor='black', lw=1.,
                        label=f'Actual ({threshold_vec[actual_threshold_id]:.2f})', zorder=3)
        ax[0].fill_between(x=fpr_vec, y1=np.zeros_like(tpr_vec), y2=tpr_vec, color='black', alpha=.05)
        ax[1].fill_between(x=tpr_vec, y1=np.zeros_like(tpr_vec), y2=pre_vec, color='black', alpha=.05)
        ax[0].set_xticks(np.arange(0, 1.01, .1))
        ax[0].set_yticks(np.arange(0, 1.01, .1))
        ax[1].set_xticks(np.arange(0, 1.01, .1))
        ax[1].set_yticks(np.arange(0, 1.01, .1))
        ax[0].set_xlim(0. - 0.01, 1. + 0.01)
        ax[0].set_ylim(0. - 0.01, 1. + 0.01)
        ax[1].set_xlim(0. - 0.01, 1. + 0.01)
        ax[1].set_ylim(y_test.mean() - .01, 1. + 0.01)
        ax[0].set_xlabel('FPR')
        ax[0].set_ylabel('TPR')
        ax[1].set_xlabel('Recall')
        ax[1].set_ylabel('Precisão')
        ax[0].set_title('ROC'+model_title+'\n$(\\text{AUROC}='+ffp(auroc_test, dig, min_digits=dig)+')$')
        ax[1].set_title('PRC'+model_title+'\n$(\\text{AUPRC}='+ffp(auprc_test, dig, min_digits=dig)+')$')
        ax[0].grid(lw=.5)
        ax[1].grid(lw=.5)
        ax[0].legend(loc='best')
        ax[1].legend(loc='best')
        plt.show()

    return metrics_dict

class BinaryRandomGuess:
    def __init__(self, threshold=0.5, seed=None):
        self.threshold = threshold
        self._uniform = sc.stats.uniform(0, 1)
        self._seed = seed

    def fit(self, X, y):
        pass

    def predict(self, X):
        n = X.shape[0]
        y_pred = self._uniform.rvs(size=n, random_state=self._seed)
        y_pred = np.int32(y_pred >= self.threshold)
        return y_pred

    def predict_proba(self, X):
        n = X.shape[0]
        y_prob_pred = self._uniform.rvs(size=n, random_state=self._seed).reshape(-1, 1)
        y_prob_pred = np.concatenate((1. - y_prob_pred, y_prob_pred), axis=1)
        return y_prob_pred

# Retorna uma função que realça as células que contém os n-ésimos valores máximos
def pd_highlight_nthmax(nth_max=1):
    def highlight_max(s, props=""):
        a = np.array([''] * np.size(s), dtype='object')
        a[np.argsort(s.values)[-nth_max]] = props
        return a
    return highlight_max
```

# Preparação do *dataset* para modelagem preditiva

```python
df = pd.read_csv('data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
display(df.head())
display(df.info())
```

```python
df = df.drop(columns='customerID')

col_missing_values = np.where(df['TotalCharges'] == ' ')[0]
df.loc[col_missing_values, 'TotalCharges'] = df.loc[col_missing_values, 'MonthlyCharges']
del col_missing_values

df = df.astype({'SeniorCitizen':'object', 'tenure':'float64', 'TotalCharges':'float64'})

display(df.info())
```

```python
df.to_csv('data/WA_Fn-UseC_-Telco-Customer-Churn_for-predictive-modelling.csv')
```

```python
df_X = df.drop(columns='Churn')
df_y = df.loc[:, ['Churn']].copy()

df_X_train, df_X_test, df_y_train, df_y_test = train_test_split(df_X, df_y, test_size=0.25, shuffle=True, stratify=df_y, random_state=7)
display(df_X_train.shape, df_X_test.shape, df_y_train.shape, df_y_test.shape)
```

```python
transformer_X = {
    'LR':ColumnTransformer(
        [('OneHot', OneHotEncoder(sparse_output=False, drop='first'), df_X.dtypes[df_X.dtypes == 'object'].index),
         ('ZScore', StandardScaler(), df_X.dtypes[df_X.dtypes == 'float64'].index)], remainder='passthrough'),
    'KNN':ColumnTransformer(
        [('OneHot', OneHotEncoder(sparse_output=False, drop='first'), df_X.dtypes[df_X.dtypes == 'object'].index),
         ('ZScore', StandardScaler(), df_X.dtypes[df_X.dtypes == 'float64'].index)], remainder='passthrough'),
    'SVM':ColumnTransformer(
        [('OneHot', OneHotEncoder(sparse_output=False, drop='first'), df_X.dtypes[df_X.dtypes == 'object'].index),
         ('ZScore', StandardScaler(), df_X.dtypes[df_X.dtypes == 'float64'].index)], remainder='passthrough'),
    'DT':ColumnTransformer(
        [('OneHot', OneHotEncoder(sparse_output=False), df_X.dtypes[df_X.dtypes == 'object'].index)], remainder='passthrough'),
    'RF':ColumnTransformer(
        [('OneHot', OneHotEncoder(sparse_output=False), df_X.dtypes[df_X.dtypes == 'object'].index)], remainder='passthrough')
}

transformer_y = OneHotEncoder(sparse_output=False, drop='if_binary')

y_train = transformer_y.fit_transform(df_y_train)
y_test = transformer_y.transform(df_y_test)
```

## Otimização de Hiperparâmetros

```python
n_comb = 1000
k_hyperparam_opt = 5

masks_kfoldcv_hyperparam_opt = list(StratifiedKFold(n_splits=k_hyperparam_opt, shuffle=True, random_state=7).split(df_X_train, df_y_train))

hyperparam_dist = {'ccp_alpha':sc.stats.loguniform(a=1e-4, b=1e3),
                   'criterion':np.random.default_rng(seed=7),
                   'min_samples_split':sc.stats.randint(2, 40+1)}
hyperparam_comb = {'ccp_alpha':hyperparam_dist['ccp_alpha'].rvs(size=n_comb, random_state=7),
                   'criterion':hyperparam_dist['criterion'].choice(['gini', 'entropy'], size=n_comb),
                   'min_samples_split':hyperparam_dist['min_samples_split'].rvs(size=n_comb, random_state=7)}

metrics_hyperparams = np.empty((5, n_comb))

time0 = perf_counter()
for i in range(k_hyperparam_opt):
    est_mask_kfoldcv, val_mask_kfoldcv = masks_kfoldcv_hyperparam_opt[i]
    
    df_X_est, df_y_est = df_X_train.iloc[est_mask_kfoldcv], df_y_train.iloc[est_mask_kfoldcv]
    df_X_val, df_y_val = df_X_train.iloc[val_mask_kfoldcv], df_y_train.iloc[val_mask_kfoldcv]

    X_est_transformed = transformer_X['Tree-Based'].fit_transform(df_X_est)
    y_est = transformer_y.fit_transform(df_y_est)
    X_val_transformed = transformer_X['Tree-Based'].transform(df_X_val)
    y_val = transformer_y.transform(df_y_val)

    for comb in tqdm(range(n_comb), desc='Validação Cruzada K-Fold - '+str(i+1)+'/5'):
        model_hyperparam_opt = DecisionTreeClassifier(ccp_alpha=hyperparam_comb['ccp_alpha'][comb],
                                                      min_samples_split=hyperparam_comb['min_samples_split'][comb],
                                                      criterion=hyperparam_comb['criterion'][comb],
                                                      random_state=7)

        model_hyperparam_opt.fit(X=X_est_transformed,
                                 y=y_est.reshape(-1))
        y_val_prob_pred = model_hyperparam_opt.predict_proba(X=X_val_transformed)[:, 1]
        
        metrics_hyperparams[i][comb] = roc_auc_score(y_score=y_val_prob_pred, y_true=y_val)
timef = perf_counter()

best_hyperparam_id = np.argmax(metrics_hyperparams.mean(axis=0))
best_hyperparam_metric = np.max(metrics_hyperparams.mean(axis=0))
best_hyperparam_comb = {'ccp_alpha':ffp(hyperparam_comb['ccp_alpha'][best_hyperparam_id]),
                        'criterion':hyperparam_comb['criterion'][best_hyperparam_id].item(),
                        'min_samples_split':ffp(hyperparam_comb['min_samples_split'][best_hyperparam_id]),
                        'auc (se)':ffp(best_hyperparam_metric, 3),
                        'Total time':ffp(timef-time0, 2)}

display(pd.DataFrame(best_hyperparam_comb, index=[0]))
```

```python
n_comb = 1000
k_hyperparam_opt = 5

masks_kfoldcv_hyperparam_opt = list(StratifiedKFold(n_splits=k_hyperparam_opt, shuffle=True, random_state=7).split(df_X_train, df_y_train))

hyperparam_dist = {'ccp_alpha':sc.stats.loguniform(a=1e-4, b=1e3),
                   'criterion':np.random.default_rng(seed=7),
                   'min_samples_split':sc.stats.randint(2, 40+1)}
hyperparam_comb = {'ccp_alpha':hyperparam_dist['ccp_alpha'].rvs(size=n_comb, random_state=7),
                   'criterion':hyperparam_dist['criterion'].choice(['gini', 'entropy'], size=n_comb),
                   'min_samples_split':hyperparam_dist['min_samples_split'].rvs(size=n_comb, random_state=7)}

def parallel_kfoldcv_fn(i):
    metrics_hyperparams = np.empty(n_comb)
    
    est_mask_kfoldcv, val_mask_kfoldcv = masks_kfoldcv_hyperparam_opt[i]
    
    df_X_est, df_y_est = df_X_train.iloc[est_mask_kfoldcv], df_y_train.iloc[est_mask_kfoldcv]
    df_X_val, df_y_val = df_X_train.iloc[val_mask_kfoldcv], df_y_train.iloc[val_mask_kfoldcv]

    X_est_transformed = transformer_X['Tree-Based'].fit_transform(df_X_est)
    y_est = transformer_y.fit_transform(df_y_est)
    X_val_transformed = transformer_X['Tree-Based'].transform(df_X_val)
    y_val = transformer_y.transform(df_y_val)

    for comb in tqdm(range(n_comb), desc='Validação Cruzada K-Fold - '+str(i+1)+'/5'):
        model_hyperparam_opt = DecisionTreeClassifier(ccp_alpha=hyperparam_comb['ccp_alpha'][comb],
                                                      min_samples_split=hyperparam_comb['min_samples_split'][comb],
                                                      criterion=hyperparam_comb['criterion'][comb],
                                                      random_state=7)

        model_hyperparam_opt.fit(X=X_est_transformed,
                                 y=y_est.reshape(-1))
        y_val_prob_pred = model_hyperparam_opt.predict_proba(X=X_val_transformed)[:, 1]
        
        metrics_hyperparams[comb] = roc_auc_score(y_score=y_val_prob_pred, y_true=y_val)

    return metrics_hyperparams

time0 = perf_counter()
metrics_hyperparams = np.array( Parallel(n_jobs=5)(delayed(parallel_kfoldcv_fn)(i) for i in range(k_hyperparam_opt)) )
timef = perf_counter()

best_hyperparam_id = np.argmax(metrics_hyperparams.mean(axis=0))
best_hyperparam_metric = np.max(metrics_hyperparams.mean(axis=0))
best_hyperparam_comb = {'ccp_alpha':ffp(hyperparam_comb['ccp_alpha'][best_hyperparam_id]),
                        'criterion':hyperparam_comb['criterion'][best_hyperparam_id].item(),
                        'min_samples_split':ffp(hyperparam_comb['min_samples_split'][best_hyperparam_id]),
                        'auc (se)':ffp(best_hyperparam_metric, 3),
                        'Total time':ffp(timef-time0, 2)}

display(pd.DataFrame(best_hyperparam_comb, index=[0]))
```

# Floresta Aleatória (**RF**)

## Otimização de Hiperparâmetros

```python
n_comb = 100
k_hyperparam_opt = 5

masks_kfoldcv_hyperparam_opt = list(StratifiedKFold(n_splits=k_hyperparam_opt, shuffle=True, random_state=7).split(df_X_train, df_y_train))

hyperparam_dist = {'n_estimators':sc.stats.randint(10, 500+1),
                   'criterion':np.random.default_rng(seed=7),
                   'max_features':sc.stats.uniform(0, 1),
                   'max_depth':np.random.default_rng(seed=7)}
hyperparam_comb = {'n_estimators':hyperparam_dist['n_estimators'].rvs(size=n_comb, random_state=7),
                   'criterion':hyperparam_dist['criterion'].choice(['gini', 'entropy'], size=n_comb),
                   'max_features':hyperparam_dist['max_features'].rvs(size=n_comb, random_state=7),
                   'max_depth':hyperparam_dist['max_depth'].choice(list(range(2, 20+1, 2)) + [None], size=n_comb)}

metrics_hyperparams = np.empty((5, n_comb))

time0 = perf_counter()
for i in range(k_hyperparam_opt):
    est_mask_kfoldcv, val_mask_kfoldcv = masks_kfoldcv_hyperparam_opt[i]
    
    df_X_est, df_y_est = df_X_train.iloc[est_mask_kfoldcv], df_y_train.iloc[est_mask_kfoldcv]
    df_X_val, df_y_val = df_X_train.iloc[val_mask_kfoldcv], df_y_train.iloc[val_mask_kfoldcv]

    X_est_transformed = transformer_X['Tree-Based'].fit_transform(df_X_est)
    y_est = transformer_y.fit_transform(df_y_est)
    X_val_transformed = transformer_X['Tree-Based'].transform(df_X_val)
    y_val = transformer_y.transform(df_y_val)

    for comb in tqdm(range(n_comb), desc='Validação Cruzada K-Fold - '+str(i+1)+'/5'):
        model_hyperparam_opt = RandomForestClassifier(n_estimators=hyperparam_comb['n_estimators'][comb],
                                                      max_features=hyperparam_comb['max_features'][comb],
                                                      criterion=hyperparam_comb['criterion'][comb],
                                                      max_depth=hyperparam_comb['max_depth'][comb],
                                                      n_jobs=-1,
                                                      random_state=7)

        model_hyperparam_opt.fit(X=X_est_transformed,
                                 y=y_est.reshape(-1))
        y_val_prob_pred = model_hyperparam_opt.predict_proba(X=X_val_transformed)[:, 1]
        
        metrics_hyperparams[i][comb] = roc_auc_score(y_score=y_val_prob_pred, y_true=y_val)
timef = perf_counter()

best_hyperparam_id = np.argmax(metrics_hyperparams.mean(axis=0))
best_hyperparam_metric = np.max(metrics_hyperparams.mean(axis=0))
best_hyperparam_comb = {'n_estimators':ffp(hyperparam_comb['n_estimators'][best_hyperparam_id]),
                        'criterion':hyperparam_comb['criterion'][best_hyperparam_id].item(),
                        'max_features':ffp(hyperparam_comb['max_features'][best_hyperparam_id]),
                        'max_depth':ffp(hyperparam_comb['max_depth'][best_hyperparam_id]),
                        'auc (se)':ffp(best_hyperparam_metric, 3),
                        'Total time':ffp(timef-time0, 2)}

display(pd.DataFrame(best_hyperparam_comb, index=[0]))
```

# Máquinas de Vetores de Suporte (**SVM**)

## Otimização de Hiperparâmetros

```python
n_comb = 100
k_hyperparam_opt = 5

masks_kfoldcv_hyperparam_opt = list(StratifiedKFold(n_splits=k_hyperparam_opt, shuffle=True, random_state=7).split(df_X_train, df_y_train))

hyperparam_dist = {'C':sc.stats.loguniform(2**-5, 2**15),
                   'gamma':sc.stats.loguniform(2**-15, 2**5)}
hyperparam_comb = {'C':hyperparam_dist['C'].rvs(size=n_comb, random_state=7),
                   'gamma':hyperparam_dist['gamma'].rvs(size=n_comb, random_state=7)}

def parallel_kfoldcv_fn(i):
    metrics_hyperparams = np.empty(n_comb)
    
    est_mask_kfoldcv, val_mask_kfoldcv = masks_kfoldcv_hyperparam_opt[i]
    
    df_X_est, df_y_est = df_X_train.iloc[est_mask_kfoldcv], df_y_train.iloc[est_mask_kfoldcv]
    df_X_val, df_y_val = df_X_train.iloc[val_mask_kfoldcv], df_y_train.iloc[val_mask_kfoldcv]

    X_est_transformed = transformer_X['LR_KNN'].fit_transform(df_X_est)
    y_est = transformer_y.fit_transform(df_y_est)
    X_val_transformed = transformer_X['LR_KNN'].transform(df_X_val)
    y_val = transformer_y.transform(df_y_val)

    for comb in range(n_comb):
        model_hyperparam_opt = SVC(C=hyperparam_comb['C'][comb],
                                   gamma=hyperparam_comb['gamma'][comb],
                                   kernel='rbf',
                                   random_state=7,
                                   probability=True)

        model_hyperparam_opt.fit(X=X_est_transformed,
                                 y=y_est.reshape(-1))
        y_val_prob_pred = model_hyperparam_opt.predict_proba(X=X_val_transformed)[:, 1]
        
        metrics_hyperparams[comb] = roc_auc_score(y_score=y_val_prob_pred, y_true=y_val)

    return metrics_hyperparams

time0 = perf_counter()
metrics_hyperparams = np.array([res for res in
                                tqdm(Parallel(n_jobs=5, return_as='generator')(
                                    delayed(parallel_kfoldcv_fn)(i) for i in range(k_hyperparam_opt)
                                ), desc=f'Validação Cruzada {k_hyperparam_opt}-fold')])
timef = perf_counter()

best_hyperparam_id = np.argmax(metrics_hyperparams.mean(axis=0))
best_hyperparam_metric = np.max(metrics_hyperparams.mean(axis=0))
best_hyperparam_comb = {'C':ffp(hyperparam_comb['C'][best_hyperparam_id]),
                        'gamma':ffp(hyperparam_comb['gamma'][best_hyperparam_id]),
                        'auc':ffp(best_hyperparam_metric, 3),
                        'Total time':ffp(timef-time0, 2)}

display(pd.DataFrame(best_hyperparam_comb, index=[0]))
```

## Ajuste de *Threshold* visando decisões de negócio

Consideremos:
- A classe de interesse maior, que são os **clientes que cometerão *churn***, é representada por **1**;
- A classe dos que **não cometerão** é representada por **0**;
- O custo de errar a classe **1**, ou seja, **classificar um cliente que é propenso a *churn* como não propenso**, é representado por $\text{c}_{10}$; 
- O custo de errar a classe **0**, ou seja, **classificar um cliente que não é propenso a *churn* como propenso**, é representado por $\text{c}_{01}$;
- O custo $\text{c}_{01}$ está relacionado à **investir na retenção de clientes que NÃO IRÃO ABANDONAR o serviço**, alocando recursos desnecessariamente;
- O custo $\text{c}_{10}$ está relacionado a **não investir na retenção de clientes IRÃO ABANDONAR o serviço**, prejudicando a receita recorrente.
- Ambos são prejudiciais, mas no contexto de negócios, **não investir na retenção de clientes que irão praticar *churn* é mais preocupante**, portanto o custo $\text{c}_{10}$ é maior.


Se um modelo retorna as probabilidades de classificação das classes **1** e **0** como $p_1$ e $p_0=1-p_1$, respectivamente, devemos escolher a classe **1** caso:

$\begin{eqnarray}
&&p_1 \cdot c_{10} > p_0 \cdot c_{01} \Rightarrow \\
&&p_1 \cdot c_{10} > (1-p_1) \cdot c_{01} \Rightarrow \\
&&p_1 > \dfrac{c_{01}}{c_{01}+c_{10}};
\end{eqnarray}$

Se considerarmos que o custo de errar a classe **1** é $R$ vezes o custo de errar a classe **0** ($c_{10}=R \cdot c_{01}$), devemos escolher a classe **1** caso:

$\begin{eqnarray}
&&p_1 > \dfrac{1}{R+1}.
\end{eqnarray}$

#### **Simulação com $R=2$ ($c_{10}$ 2 vezes maior do que $c_{01}$)**
> 
Vamos considerar:
>O **custo de não reter clientes que irão abandonar o serviço** é **2x** maior do que o **custo de investir desnecessariamente em clientes que não abandonariam**;

Isso significa que devemos escolher a classe **1** caso $p_1 > 1/3$, ou seja, **o novo *threshold* será 0.33.**

#### Observações da comparação final:
- Como as ponderações para os diferentes tipos de predições já estão sendo levadas em conta a partir dos diferentes custos de erros, a ponderação pela métrica $\text{F}_{\beta}\text{ -Score}$ ($\beta=2$) agora faz menos sentido;
- O modelo **LR** apresentou o melhor equilíbrio entre precisão e *recall*;
- O modelo **SVM** cresceu significamente em *recall*, ao custo de ter sua precisão reduzida também significativamente;
- Os modelos **DT** e **KNN** apresentaram-se competitivos, com **DT** priorizando precisão, e **KNN** priorizando *recall*;
- O modelo **RF** melhorou o equilíbrio entre precisão e *recall* com relação ao **DT**, mas ainda esteve atrás do **LR**;

#### **Conclusão:**
> O modelo **LR** deve ser escolhido: Apresenta melhor $\text{F}_1\text{ - Score}$, é computacionalmente leve e um dos mais interpretáveis.
>
> **Impacto da Escolha:**
> - 75% de *churns* identificados;
> - 42% de clientes impactados desnecessariamente;

- O modelo **RF** melhorou o equilíbrio entre precisão e *recall* com relação ao **DT**, mas ainda esteve atrás do **LR**;

#### **Conclusão:**
> O modelo **LR** deve ser escolhido: Apresenta melhor $\text{F}_1\text{ - Score}$, é computacionalmente leve e um dos mais interpretáveis.
>
> **Impacto da Escolha:**
> - 96% de *churns* identificados;
> - 60% de clientes impactados desnecessariamente;
